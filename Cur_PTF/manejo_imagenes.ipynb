```python
arr = [11, 21, 7, 14, 3, 6, 6, 6, 1, 21]

def indice_mayor_a_menor_en_array(arr):

    contados = set()
    arr_sin_repetir = []
    
    for value in arr:
        if value not in contados:
            arr_sin_repetir.append(value)
            contados.add(value)

    dic = {value: key + 1 for key, value in enumerate(sorted(arr_sin_repetir, reverse=True))}

    solucion = [dic[value] for value in arr]
    
    return solucion

print(indice_mayor_a_menor_en_array(arr))
```
